### Importing the Necessary Libraries

```python
from bs4 import BeautifulSoup

import requests
import csv
```

### Finding all the 15 pages of the AirBnB Paris Location for 1st Dec to 31st Dec 2020 and for 2 Guests

```python
#Defining my main URL
URL = 'https://www.airbnb.com/s/Paris--France/homes?adults=2&source=structured_search_input_header&search_type=filter_change&tab_id=home_tab&checkin=2020-12-01&refinement_paths%5B%5D=%2Fhomes&_set_bev_on_new_domain=1605324153_6EZWMnMrkGClx6tR&checkout=2020-12-31&place_id=ChIJD7fiBh9u5kcRYJSMaMOCCwQ&locale=en'

noOfPages = 15 #The airbnb website has a total of 15 pages that we need to loop to


urlList = [] # Defining an empty list that will contain all the 15 links of all 15 pages

#Using a for loop to append each page's link to the urllist List
for i in range(1, noOfPages):
    urlApp = URL+'&items_offset='+str(i*20)+'&section_offset=3'
    print(urlApp)
    urlList.append(urlApp)
```

### The Main Scraping using BeautifulSoup and writing to a CSV File

```python
#For each of the URL from the urlList, we need to save the details:
for URL in urlList:
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    results = soup.find(class_='_fhph4u') #This saves the main page that contains all the hotel details of that page
    #print(results.prettify())
    stay_elems = results.find_all('div', class_='_gig1e7')
    for stay_elem in stay_elems: #Looping through all hotels in the above main page
        place = stay_elem.find(class_='_167qordg').text #Scraping the place of the first hotel and so on
        addrs = stay_elem.find(class_='_bzh5lkq').text #Scraping the address of the first hotel and so on


        descElemt = stay_elem.find_all(class_='_kqh46o') #Scraping the description of the first hotel and so on
        descText = ''
        
        for desc in descElemt:
            descText+=desc.text + '. '
        #print(descText)
        
        try:
            ratingsElement = stay_elem.find(class_='_18khxk1') #Scraping the ratings of the first hotel and so on
            #print(ratingsElement.find(class_='_krjbj').text)
            ratings = ratingsElement.find(class_='_10fy1f8').text #Scraping the ratings of the first hotel and so on
            #print(ratings)
            reviews = ratingsElement.find_all(class_='_krjbj') #Scraping the reviews of the first hotel and so on
            revText=''
            for rev in reviews: 
                #print(rev.text)
                revText+=rev.text+' '
            #print(revText)
        except Exception as e: #If no ratings are present
            #print('No Ratings Available')
            revText='No Ratings Available'
            ratings='No Ratings Available'

        priceElement = stay_elem.find(class_='_l2ulkt8')
        price = priceElement.find(class_='_1p7iugi').text.replace('Discounted', '; Discounted')
        print(place)
        print(addrs)
        print(descText)
        print(ratings)
        print(revText)
        print(price)
        
        #Saving the first scraped hotel to the excel file by appending every row and so on
        with open(r"C:\Users\ktirumalelakshmana\Desktop\hotels.csv", 'a', newline='', encoding="utf-8") as csvfile:
            spamwriter = csv.writer(csvfile, delimiter=',',
                                quotechar='"', quoting=csv.QUOTE_MINIMAL)
            spamwriter.writerow([place,addrs,ratings,revText,price] + descText.split(' · ') )
        print('\n')

```
